```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
merged_df = pd.read_csv("data/data_merged.csv")
```

```python
merged_df = merged_df[merged_df["questionnaire_login_year_group_name"].isin(["Year 9", "Year 10", "Year 11", "Year 12"])]
```

```python
[i for i in list(merged_df.columns) if "dining" in i]
```

```python
merged_df["How much do you agree or disagree with the following statements?_I enjoy eating in my school/college dining area"].value_counts(normalize=True).plot.bar()
```

```python
enjoy_eating_in_dining = merged_df["How much do you agree or disagree with the following statements?_I enjoy eating in my school/college dining area"]
```

```python
enjoy_eating_in_dining[enjoy_eating_in_dining != "Agree"] = "Not Agree"
```

```python
enjoy_eating_in_dining.value_counts(normalize=True).sort_values().plot(kind="bar", label="index", colormap="Paired")
```

```python
students_behave_well = merged_df[enjoy_eating_in_dining!="Agree"]["How much do you agree or disagree with the following statements?_Pupils/students behave well in the dining area"]
```

```python
students_behave_well[students_behave_well != "Agree"] = "Do not Agree"
```

```python
students_behave_well.value_counts(normalize=True).plot(kind="pie", autopct="%1.1f%%", title="Students that agree that students behave well in the dining area")
```

```python
merged_df[merged_df["How much do you agree or disagree with the following statements?_I enjoy eating in my school/college dining area"]!="Agree"]["How much do you agree or disagree with the following statements?_Pupils/students behave well in the dining area"].value_counts(normalize=True).plot.bar()
```
