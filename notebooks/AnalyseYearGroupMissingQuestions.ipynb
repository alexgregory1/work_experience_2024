```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw_df = pd.read_csv("data/data_raw.csv", header=[0, 1], low_memory=False)
```

```python
raw_df
```

```python
merged_data = pd.read_csv("data/data_merged.csv", low_memory=False)
```

```python
merged_data.head()
```

```python
pct_missing_questions_by_year_group = merged_data.groupby("questionnaire_login_year_group_name").apply(lambda x: x.isna().sum() / x.shape[0])
```

```python
pct_missing_questions_by_year_group
```

```python
(pct_missing_questions_by_year_group < 0.5).sum(axis=1)
```

```python
plt.title("Students grouped by their year (Answered above 50% of questions)")
(pct_missing_questions_by_year_group.filter(regex="^Year", axis=0) < 0.5).sum(axis=1).sort_values().plot.bar()
```

### Conclusion:
- There are some questions that are inappropriate for certain year groups (Lots of empty responses for year 3 & 4)
- The number of questions answered across year 9 - 12 stayed constant
