```python
# strategy is, if header 1 is "Unnamed", then the new header is the last 

new_dataframe = {}

combined_headers = []

last_h1_header = None

for h1, h2 in zip(header_1, header_2):
    if h1.startswith("Unnamed"):
        if last_h1_header:
            combined_headers.append(f"{last_h1_header}_{h2}")
        else:
            combined_headers.append(h2)
        
    else:
        last_h1_header = h1
        
    new_dataframe[combined_headers[-1]] = df[h1, h2]
```

```python
new_df = pd.DataFrame(new_dataframe)
```
